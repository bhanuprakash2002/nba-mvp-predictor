```python
from IPython.display import display
import pandas

from nba_mvp_predictor import load
```

```python
pandas.options.display.float_format = '{:,.1%}'.format
```

```python
stats = load.load_player_stats(nrows=None)
stats = stats.rename(columns={'EFG%_totals':'eFG%'})
new_col = 'Player'.ljust(13) +  '(season)'
stats[new_col] = '' + stats['PLAYER'].str.ljust(15) + ' (' + stats['SEASON'].astype(str) +')'
#stats = stats.set_index(new_col, drop=True)
#stats = stats.set_index(['PLAYER', 'SEASON', 'TEAM'], drop=True)
stats.index.name = None
stats = stats.sort_values(by='SEASON', ascending=True)
```

* 3PT %:    82 three-point field goals made
* FIELD GOAL %:    300 field goals made
* FREE THROW %:    125 free throws made

```python
eligible_seasons = stats[
    (stats['3P_totals'] >= 72) #2022 : 82 games. Set to 72 for Dirk
    & (stats['FG_totals'] >= 300)
    & (stats['FT_totals'] >= 125)
]
len(eligible_seasons)
```

```python
original_cols = [
    new_col,
    'FG%', '3P%', 'FT%'
]

cols = [
    new_col,
    '50-40-90',
    '2P%', '3P%', 'FT%',
]
```

```python
eligible_seasons['50-40-90'] = (eligible_seasons['3P%'] >= 0.400) & (eligible_seasons['FG%'] >= 0.500) & (eligible_seasons['FT%'] >= 0.900)
eligible_seasons['50-40-90'] = eligible_seasons['50-40-90'].map({True:'Yes', False:'No'})
```

```python
club = eligible_seasons[eligible_seasons['50-40-90'] == 'Yes'][original_cols]
display(club)
```

```python
new_club = eligible_seasons[
    (eligible_seasons['3P%'] >= 0.400)
    & (eligible_seasons['2P%'] >= 0.500)
    & (eligible_seasons['FT%'] >= 0.900)
][cols]
cols_new_members = cols
new_members = new_club[new_club['50-40-90'] == 'No']
cols_new_members.remove("50-40-90")
display(new_members[cols_new_members])
```
